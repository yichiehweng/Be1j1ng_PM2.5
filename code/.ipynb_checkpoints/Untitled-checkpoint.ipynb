```r
require(MASS)
```

```r
df=read.csv("/home/yichiehweng/python/machineLearning/ridge_regression/Beijing_PM25/data/PRSA_data.csv")
```

```r
head(df,5)
```

```r
sapply(df, function(x) sum(is.na(x)))      
```

```r
f=function(x){

  x<-as.numeric(as.character(x)) 

  x[is.na(x)] =median(x, na.rm=TRUE) 

  x

}
```

```r
df["pm2.5"]<-data.frame(apply(df["pm2.5"],2,f))
```

```r
df["pm"]<-df["pm2.5"]
```

```r
model_1 <- lm(pm~ year+month+day+hour+DEWP+TEMP+PRES+cbwd+Iws+Is+Ir,df)
```

```r
summary(model_1)
```

```r
ridge <- lm.ridge(pm~ year+month+day+hour+DEWP+TEMP+PRES+cbwd+Iws+Is+Ir,df, lambda = seq(0, .1, .001))
```

```r
summary(ridge)
```

```r
plot(ridge)
```

```r
select(ridge)
```

```r
ridge_2 <- lm.ridge(pm~ year+month+day+hour+DEWP+TEMP+PRES+cbwd+Iws+Is+Ir,df, lambda = seq(0, 1, .1))
plot(ridge_2)
select(ridge_2)
```

```r
ridge_3 <- lm.ridge(pm~ year+month+day+hour+DEWP+TEMP+PRES+cbwd+Iws+Is+Ir,df, lambda = seq(0, 10, .01))
plot(ridge_3)
select(ridge_3)
```
